# Basic Anomaly Detection

*   Based on Median Absolute Deviation

*   One of the commonest ways of finding outliers in one-dimensional data is to mark as a potential outlier any point that is more than two standard deviations (sample means and standard deviations here and in what follows)  
[Using the Median Absolute Deviation to Find Outliers](https://eurekastatistics.com/using-the-median-absolute-deviation-to-find-outliers/)









```
import numpy as np
import matplotlib.pyplot as mplt
```

```
#Function to detect outlier
def detect_outliers(data): 
    # mean and standard deviation 
    u = np.mean(data)
    s = np.std(data)
    filtered = []
    # the points on graph
    for e in data:
        if (u - 2 * s < e < u + 2 * s):            
            filtered.append(e)
        else:
            mplt.plot(e,'ro')
    mplt.plot(filtered,'go')
    return filtered
```

```
#Testcase 1
list1 = [101,201,301,50,100,20,30,40,50,90,10,400000,500000,600000]
filtered_d = detect_outliers(list1)
mplt.title("Testcase 1")
mplt.show()
```

```
#Testcase 2
list2 = [1,3,5,7,6,11,13,15,40,20,10,12,32,80,10,9,5,3,50,300,200,400,500]
filtered_d = detect_outliers(list2)
mplt.title("Testcase 2")
mplt.show()
```

```
#Testcase 3
list3 = [1, 2, 3, 3, 4, 4, 4, 5, 5.5, 6, 6, 6.5, 7, 7, 7.5, 8, 9, 12, 52, 90]
filtered_d = detect_outliers(list3)
mplt.title("Testcase 3")
mplt.show()
```
